```python
import seaborn as sns
df=sns.load_dataset("titanic")
df.head()
```

```python
import plotly.express as px
fig=px.scatter(df,x='age',y='fare',color='class')
fig.show()
```

```python
df.columns
```

```python
class_counts = df['pclass'].value_counts().reset_index()
class_counts.columns = ['Class', 'Count']

fig = px.bar(class_counts, x='Class', y='Count', text='Count', title='Passengers by Class')
fig.update_traces(texttemplate='%{text}', textposition='inside')
fig.show()
```

```python
df.columns
```

```python
# Load the Titanic dataset
import seaborn as sns
df=sns.load_dataset('titanic')
gender_counts = df['sex'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']

fig = px.pie(gender_counts, names='Gender', values='Count', title='Gender Distribution')
fig.show()
```

```python
# Load the Titanic dataset
import seaborn as sns
df=sns.load_dataset('titanic')
gender_counts = df['survived'].value_counts().reset_index()
gender_counts.columns = ['survived', 'Count']

fig = px.pie(gender_counts, names='survived', values='Count', title='survived Distribution')
fig.show()
```

```python
fig = px.histogram(df, x='age', nbins=20, title='Age Distribution')
fig.show()
```

```python
fig = px.box(df, x='pclass', y='fare', title='Fare Distribution by Class')
fig.show()
```
